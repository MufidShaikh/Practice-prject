Separating the Columns into Features and Target

```python
features=df.drop('Rings',axis=1)
target=df['Rings']
```

```python
features.skew()
```

```python
from sklearn.preprocessing import power_transform
features_new=power_transform(features)
features=pd.DataFrame(features_new,columns=features.columns)
features.skew()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
lr=LinearRegression()
```

```python
maxAcc=0
maxRS=0
for i in range(0,100):
    features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=.20,random_state=i)
    lr.fit(features_train,target_train)
    pred_train=lr.predict(features_train)
    pred_test=lr.predict(features_test)
    print(f"At random state {i},the training accuracy is:- {r2_score(target_train,pred_train)}")
    print(f"At random state {i},the testing accuracy is:- {r2_score(target_test,pred_test)}")
    print('\n')
    acc=r2_score(target_test,pred_test)
    if(acc>maxAcc):
        maxAcc=acc
        maxRS=i
print("Best accuracy: ",maxAcc," on Random State: ",maxRS)
```

```python
features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=.20,random_state=43)
```

```python
lr.fit(features_train,target_train)
```

```python
pred_test=lr.predict(features_test)
```

```python
print(r2_score(target_test,pred_test))
```

# Cross-validation:

```python
train_acc=r2_score(target_train,pred_train)
test_acc=r2_score(target_test,pred_test)
from sklearn.model_selection import cross_val_score
for j in range(2,10):
    cv_score=cross_val_score(lr,features,target,cv=j)
    cv_mean=cv_score.mean()
    print(f"At Cross fold {j} cv score:{cv_mean} Accuracy Training: {train_acc} Accuracy Testing:{test_acc}")
    print('\n')
```
